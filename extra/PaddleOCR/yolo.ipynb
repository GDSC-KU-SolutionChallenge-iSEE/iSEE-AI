```python
import cv2
import numpy as np
import cv2
import numpy as np
from ultralytics import YOLO

from ultralytics.utils.checks import check_imshow
from ultralytics.utils.plotting import Annotator, colors

from collections import defaultdict
from google.colab.patches import cv2_imshow
import time

track_history = defaultdict(lambda: [])
# Load the image
image_path = '/content/drive/MyDrive/iSEE/ultralytics/bus.jpg'  # Update this path to your image file
frame = cv2.imread(image_path)

# import pdb;pdb.set_trace()
# Assuming 'model' is already loaded and ready for inference
start_time = time.time()

results = model("https://ultralytics.com/images/bus.jpg")
# results = model.track(frame, persist=True, verbose=False)
end_time = time.time()

# Calculate the inference time
inference_time = end_time - start_time

print(f"Model inference time: {inference_time:.3f} seconds")


names = model.model.names
boxes = results[0].boxes.xyxy.cpu()


# import pdb;pdb.set_trace()

if results[0].boxes.id is not None:

    # Extract prediction results
    clss = results[0].boxes.cls.cpu().tolist()
    track_ids = results[0].boxes.id.int().cpu().tolist()
    confs = results[0].boxes.conf.float().cpu().tolist()

    # Annotator Init (assuming Annotator class is already defined)
    annotator = Annotator(frame, line_width=2)

    for box, cls, track_id in zip(boxes, clss, track_ids):
        annotator.box_label(box, color=colors(int(cls), True), label=names[int(cls)])  # Assuming 'colors' and 'names' functions are defined

        # Store tracking history (assuming 'track_history' dict is defined)
        track = track_history.get(track_id, [])
        track.append((int((box[0] + box[2]) / 2), int((box[1] + box[3]) / 2)))
        if len(track) > 30:
            track.pop(0)
        track_history[track_id] = track

        # Plot tracks
        points = np.array(track, dtype=np.int32).reshape((-1, 1, 2))
        cv2.circle(frame, track[-1], 7, colors(int(cls), True), -1)
        cv2.polylines(frame, [points], isClosed=False, color=colors(int(cls), True), thickness=2)

# Save or display the annotated image
cv2.imwrite('/content/drive/MyDrive/iSEE/ultralytics/result.jpg', frame)  # Update this path to where you want to save the annotated image
# or display the image in a window
cv2_imshow(frame)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from ultralytics import YOLO

from ultralytics.utils.checks import check_imshow
from ultralytics.utils.plotting import Annotator, colors

from collections import defaultdict

track_history = defaultdict(lambda: [])
model = YOLO("yolov8n.pt")
names = model.model.names

video_path = "/path/to/video/file.mp4"
cap = cv2.VideoCapture(video_path)
assert cap.isOpened(), "Error reading video file"

w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

result = cv2.VideoWriter("object_tracking.avi",
                       cv2.VideoWriter_fourcc(*'mp4v'),
                       fps,
                       (w, h))

while cap.isOpened():
    success, frame = cap.read()
    if success:
        results = model.track(frame, persist=True, verbose=False)
        boxes = results[0].boxes.xyxy.cpu()

        if results[0].boxes.id is not None:

            # Extract prediction results
            clss = results[0].boxes.cls.cpu().tolist()
            track_ids = results[0].boxes.id.int().cpu().tolist()
            confs = results[0].boxes.conf.float().cpu().tolist()

            # Annotator Init
            annotator = Annotator(frame, line_width=2)

            for box, cls, track_id in zip(boxes, clss, track_ids):
                annotator.box_label(box, color=colors(int(cls), True), label=names[int(cls)])

                # Store tracking history
                track = track_history[track_id]
                track.append((int((box[0] + box[2]) / 2), int((box[1] + box[3]) / 2)))
                if len(track) > 30:
                    track.pop(0)

                # Plot tracks
                points = np.array(track, dtype=np.int32).reshape((-1, 1, 2))
                cv2.circle(frame, (track[-1]), 7, colors(int(cls), True), -1)
                cv2.polylines(frame, [points], isClosed=False, color=colors(int(cls), True), thickness=2)

        result.write(frame)
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break

result.release()
cap.release()
cv2.destroyAllWindows()
```
